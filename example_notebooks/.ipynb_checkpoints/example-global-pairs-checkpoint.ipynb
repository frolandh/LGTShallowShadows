```python
import numpy as np

from src.basis import calculate_z2_basis, calculate_ising_basis, getH_Z2, getH_ising, z2_to_ising
from src.measurements import run_experiment
from src.allpairs import compute_sample_pairs, get_z, get_xy, get_betas,compute_sample_ising_z
from tqdm import tqdm
import matplotlib.pyplot as plt
from scipy import sparse

from quspin.basis import spin_basis_1d
```

### Setup System

Run Global Pairs on a 2x2 lattice

```python
Nx,Ny = 3,2
Vx,Vy = -1,1
size=[Nx,Ny]
parity = [Vx,Vy]

g=1.0
K_list = [-5.0,-3.0,0.0,3.0,5.0]
coupling = np.zeros(len(K_list),dtype=object)
K_ctr = 0
for K in K_list:
    coupling[K_ctr] = [K,g]
    K_ctr += 1
```

```python
#Create 2 and dual bases
basis_z2 = calculate_z2_basis(size,parity)
basis_z2_full = spin_basis_1d(2*Nx*Ny)
basis_ising = calculate_ising_basis(size)
basis_ising_full = spin_basis_1d(Nx*Ny)
```

```python
#Create hamiltonians for scan
H_z2_list = np.zeros(len(K_list),dtype=object)
H_is_list = np.zeros(len(K_list),dtype=object)
h_ctr = 0
for c in coupling:
    H_z2_list[h_ctr] = getH_Z2(size,c,basis_z2)
    H_is_list[h_ctr] = getH_ising(size,c,parity,basis_ising)

    h_ctr += 1
```

```python
#Test ground state energies are equal
E0_Z2,psi0_Z2 = H_z2_list[1].eigsh(k=1,which="SA")
E0_ising, psi0_ising = H_is_list[1].eigsh(k=1,which="SA")

psi0_z2_all = np.zeros(len(K_list),dtype=object)
psi0_ising_all = np.zeros(len(K_list),dtype=object)

psi0_z2_full = np.zeros(len(K_list),dtype=object)
psi0_ising_full = np.zeros(len(K_list),dtype=object)
s_ctr = 0
for i_h in range(len(H_is_list)):
    _,psi0_z2 = H_z2_list[i_h].eigsh(k=1,which="SA")
    psi0_z2_all[s_ctr] = psi0_z2
    
    _,psi0_ising = H_is_list[i_h].eigsh(k=1,which="SA")
    psi0_ising_all[s_ctr] =  psi0_ising

    s_ctr += 1

print(f"Z2 ground energy - {E0_Z2}")
print(f"Ising ground energy - {E0_ising}")
```

### Run Global Pairs Protocol

```python
num_shots = int(1e3)

table_list = np.zeros(len(psi0_z2_all),dtype=object)

for i_s, state in enumerate(psi0_z2_all):
    shadow_table = run_experiment(
        state,
        num_shots,
        size,
        basis_z2,
        basis_z2_full,
        "pairs"
    )

    table_list[i_s] = shadow_table
```

### Reconstruct Loop and String Observables

Calculates loop first and then string

We use sample mean and standard error from the mean for a single experiment

```python
all_samples_z = np.zeros(len(K_list),dtype=object)

#Test operator to reconstruct
z_str_ising = [1,0,1,0,0,0,0,0]
x_str_ising = [0,0,0,0,0,0,0,0]
y_str_ising = [0,0,0,0,0,0,0,0]

samp_ctr = 0
for shadow_table in table_list:
    sample_list = []

    for entry in tqdm(shadow_table,desc="Computing Samples"):
        angles,pairs,b = entry
        sample = compute_sample_pairs(
            x_str_ising,
            y_str_ising,
            z_str_ising,
            b,
            angles,
            pairs,
            size,
            basis_ising,
            basis_ising_full,
            basis_z2_full
        )
            
        sample_list.append(sample)

    all_samples_z[samp_ctr] = sample_list
    samp_ctr += 1
```

```python
est_list_loop = np.zeros(len(K_list),dtype=object)
err_list_loop = np.zeros(len(K_list),dtype=object)

samp_ctr = 0
for sample_list in all_samples_z:
    est = np.mean(sample_list)
    err = np.std(sample_list) / np.sqrt(num_shots)

    est_list_loop[samp_ctr] = est
    err_list_loop[samp_ctr] = err

    samp_ctr += 1
```

```python
#Calculate exact value for comparison

loop_op_is = get_z(z_str_ising,basis_ising_full)
ext_list_loop = np.zeros(len(K_list),dtype=object)

ext_ctr = 0
for state in psi0_ising_all:
    state_full = basis_ising.project_from(state)
    if type(loop_op_is) != sparse._csc.csc_matrix:
        exact = loop_op_is.expt_value(state_full)
        
    ext_list_loop[ext_ctr] = np.real(exact.item())

    ext_ctr += 1
    
```

```python
disp_str = "".join([f"Z_{i}" if z_str_ising[i] != 0 else "" for i in range(len(x_str_ising))])
disp_str = "<" + disp_str + ">"

fig, axs = plt.subplots(1,2,figsize=(15,5))

axs[0].errorbar(K_list,est_list_loop,yerr=err_list_loop,marker='o',label='estimate',ls='',c='b')
axs[0].scatter(K_list,ext_list_loop,label='exact',c='orange')
axs[0].legend(frameon=False)
axs[0].set_ylabel(disp_str)
axs[0].set_title(f"Expectation value vs. K ({num_shots} samples, Z2 measure ising reconstruction)")

axs[1].errorbar(K_list,ext_list_loop-est_list_loop,yerr=err_list_loop,marker='o',label='estimate',ls='',c='b')
axs[1].set_ylabel("residual")
plt.show()
```

```python
all_samples_xy = np.zeros(len(K_list),dtype=object)

#Test string to reconstruct
z_str_ising = [0,0,0,0,0,0,0,0]
x_str_ising = [1,0,1,0,0,0,0,0]
y_str_ising = [0,0,0,0,0,0,0,0]

samp_ctr = 0
for shadow_table in table_list:
    sample_list = []

    for entry in tqdm(shadow_table,desc="Computing Samples"):
        angles,pairs,b = entry
        sample = compute_sample_pairs(
            x_str_ising,
            y_str_ising,
            z_str_ising,
            b,
            angles,
            pairs,
            size,
            basis_ising,
            basis_ising_full,
            basis_z2_full
        )
            
        sample_list.append(sample)

    all_samples_xy[samp_ctr] = sample_list
    samp_ctr += 1
```

```python
#reconstruct observable
est_list_string = np.zeros(len(K_list),dtype=object)
err_list_string = np.zeros(len(K_list),dtype=object)

samp_ctr = 0
for sample_list in all_samples_xy:
    est = np.mean(sample_list)
    err = np.std(sample_list) / np.sqrt(num_shots)

    est_list_string[samp_ctr] = est
    err_list_string[samp_ctr] = err

    samp_ctr += 1
```

```python
#Calculate the exact for comparison
string_op = get_xy(x_str_ising,y_str_ising,basis_ising_full)

ext_list_string = np.zeros(len(K_list),dtype=object)

ext_ctr = 0
for state in psi0_ising_all:
    state_full = basis_ising.project_from(state)
    if type(string_op) != sparse._csc.csc_matrix:
        exact = string_op.expt_value(state_full)
        
    ext_list_string[ext_ctr] = np.real(exact.item())

    ext_ctr += 1
```

```python
disp_str = [f"X_{i}" if x_str_ising[i] != 0 else "" for i in range(len(x_str_ising))]
disp_str = [disp_str[i] + f"Y_{i}" if y_str_ising[i] != 0 else disp_str[i] for i in range(len(y_str_ising))]
disp_str = "".join(disp_str)
disp_str = "<" + disp_str + ">"

fig, axs = plt.subplots(1,2,figsize=(15,5))

axs[0].errorbar(K_list,est_list_string,yerr=err_list_string,marker='o',label='estimate',ls='',c='b')
axs[0].scatter(K_list,ext_list_string,label='exact',c='orange')
axs[0].legend(frameon=False)
axs[0].set_xlabel("K (g=1)")
axs[0].set_ylabel(disp_str)
axs[0].set_title(f"Expectation value vs. K ({num_shots} samples, Z2 measure ising reconstruction)")

axs[1].errorbar(K_list,ext_list_string-est_list_string,yerr=err_list_string,marker='o',label='estimate',ls='',c='b')
axs[1].set_ylabel("residual")

plt.show()
```
